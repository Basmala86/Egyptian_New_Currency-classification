```python
import os
import zipfile
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from sklearn.metrics import classification_report, confusion_matrix
from google.colab import files

# ==========================================
# 0. GPU CHECK & SETUP
# ==========================================
print(f"[INFO] TensorFlow Version: {tf.__version__}")
gpu_devices = tf.config.list_physical_devices('GPU')
if gpu_devices:
    print(f"[SUCCESS] Running on GPU: {gpu_devices[0].name}")
else:
    print("[WARNING] Running on CPU! Go to Runtime > Change runtime type > T4 GPU")

# ==========================================
# 1. DATASET DOWNLOAD & EXTRACTION
# ==========================================
# Remove old data if exists to prevent conflicts
if os.path.exists('egyptian-new-currency-2023.zip'):
    os.remove('egyptian-new-currency-2023.zip')

print("\n[STEP 1] Upload kaggle.json file...")
if not os.path.exists('kaggle.json'):
    files.upload() # Trigger upload dialog

# Setup Kaggle config
os.system('mkdir -p ~/.kaggle')
os.system('cp kaggle.json ~/.kaggle/')
os.system('chmod 600 ~/.kaggle/kaggle.json')

print("[STEP 2] Downloading Dataset...")
os.system('kaggle datasets download -d belalsafy/egyptian-new-currency-2023')

print("[STEP 3] Unzipping...")
with zipfile.ZipFile('egyptian-new-currency-2023.zip', 'r') as zip_ref:
    zip_ref.extractall('dataset_root')

# ==========================================
# 2. PATH CORRECTION (CRITICAL FIX)
# ==========================================
# Find where the actual image folders are hidden
base_path = 'dataset_root'
if 'dataset' in os.listdir(base_path):
    base_path = os.path.join(base_path, 'dataset')

# Define specific sub-directories
train_dir = os.path.join(base_path, 'train')
val_dir = os.path.join(base_path, 'validation')

# Fallback: some versions of the dataset name it 'valid' or 'val'
if not os.path.exists(val_dir):
    if os.path.exists(os.path.join(base_path, 'valid')):
        val_dir = os.path.join(base_path, 'valid')
    elif os.path.exists(os.path.join(base_path, 'val')):
        val_dir = os.path.join(base_path, 'val')

print(f"\n[INFO] Train Directory: {train_dir}")
print(f"[INFO] Val Directory:   {val_dir}")

# ==========================================
# 3. DATA LOADERS & AUGMENTATION
# ==========================================
IMG_SIZE = (128, 128) # 128x128 strikes balance between speed and detail
BATCH_SIZE = 32

# Aggressive Augmentation to replace Pre-training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.15,
    height_shift_range=0.15,
    shear_range=0.15,
    zoom_range=0.15,
    brightness_range=[0.7, 1.3], # Critical for shiny plastic notes
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

print("\n[INFO] Loading Generators...")
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

validation_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False # Important: No shuffle for confusion matrix
)

# VERIFY CLASSES (Expect 8 or more, NOT 3)
num_classes = train_generator.num_classes
class_names = list(train_generator.class_indices.keys())
print(f"\n[CHECK] Detected {num_classes} Classes: {class_names}")

# ==========================================
# 4. CUSTOM "MINI-VGG" ARCHITECTURE
# ==========================================
model = Sequential()

# Block 1: Edges & Colors
model.add(Conv2D(32, (3, 3), padding='same', input_shape=(128, 128, 3)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

# Block 2: Shapes & Patterns
model.add(Conv2D(64, (3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Conv2D(64, (3, 3), padding='same')) # Double Conv for better learning
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.3))

# Block 3: Complex Features (Watermarks/Text)
model.add(Conv2D(128, (3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Conv2D(128, (3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.4))

# Block 4: High Level Features
model.add(Conv2D(256, (3, 3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.4))

# Classification Head
model.add(Flatten())
model.add(Dense(512))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5)) # Heavy dropout to force generalization
model.add(Dense(num_classes, activation='softmax'))

# Compile
opt = Adam(learning_rate=0.001) # Start with standard LR
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# ==========================================
# 5. TRAINING
# ==========================================
# Callbacks are the secret to high accuracy
callbacks = [
    ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=3, verbose=1, min_lr=0.00001),
    EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True)
]

print("\n[INFO] Starting Training...")
history = model.fit(
    train_generator,
    epochs=50, # Set high, EarlyStopping will handle the rest
    validation_data=validation_generator,
    callbacks=callbacks
)

# ==========================================
# 6. RESULTS & VISUALIZATION
# ==========================================
# A. Plotting
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title('Loss')
plt.legend()
plt.show()

# B. Confusion Matrix
print("\n[INFO] Generating Confusion Matrix...")
Y_pred = model.predict(validation_generator)
y_pred = np.argmax(Y_pred, axis=1)
y_true = validation_generator.classes

plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix')
plt.show()

# C. Classification Report
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

# D. Save Model
model.save('egyptian_money_custom_cnn.h5')
print("[INFO] Model saved as 'egyptian_money_custom_cnn.h5'")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_predictions(generator, model, num_images=9):
    # 1. Get a batch of images and labels from the validation set
    images, labels = next(generator)

    # 2. Make predictions on this batch
    predictions = model.predict(images)

    # 3. Get Class Names (Labels)
    class_names = list(generator.class_indices.keys())

    # 4. Setup the Plot
    plt.figure(figsize=(15, 10))

    for i in range(num_images):
        plt.subplot(3, 3, i + 1)

        # Display the image
        plt.imshow(images[i])

        # Get True and Predicted Labels
        true_label_index = np.argmax(labels[i])
        pred_label_index = np.argmax(predictions[i])

        true_label = class_names[true_label_index]
        pred_label = class_names[pred_label_index]
        confidence = np.max(predictions[i]) * 100

        # Color Code: Green if Correct, Red if Wrong
        color = 'green' if true_label == pred_label else 'red'

        # Title with details
        plt.title(f"True: {true_label}\nPred: {pred_label} ({confidence:.1f}%)",
                  color=color, fontsize=12, fontweight='bold')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Run the function
print("[INFO] Generating Prediction Grid...")
plot_predictions(validation_generator, model, num_images=9)
```
